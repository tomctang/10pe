```python
# Draw Two-Worlds
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
edges_left = [("Coding(n>1)", "Maths(n=1)"), ("Maths(n=1)", "Natural_Language(n>1)"), ("Natural_Language(n>1)", "Coding(n>1)"),
            ("Maths(n=1)", "Coding(n>1)"), ("Natural_Language(n>1)","Maths(n=1)"), ("Coding(n>1)", "Natural_Language(n>1)")]
edges_right = [("Nature","Human"),("Human","Nature")]
G.add_edges_from(edges_left)
G.add_edges_from(edges_right)

# Define node positions manually
pos = {"Coding(n>1)": (-1, -0.6), "Maths(n=1)": (-3, -0.6), "Natural_Language(n>1)": (-2, 0.7),
        "Nature": (1, -0.1), "Human": (3, -0.1), }
pos["Human Information World"] = (-2, 2)  
pos["Real World"] = (3, 2)  

# Draw the graph
plt.figure(figsize=(8, 5))
nx.draw(G, pos, with_labels=True, node_size=6000, node_color="lightblue", edge_color="black", font_size=12, font_weight="bold")
plt.text(-2, 0, "Human Information World", fontsize=14, ha="center", fontweight="bold") # Add titles 
plt.text(2, 0, "Real World", fontsize=14, ha="center", fontweight="bold")
plt.axvline(x=0, color="black", linestyle="-", linewidth=1) # Draw a vertical line
plt.show()
```

In this question, I learned Ordinal theory (VS. Cardinal theory):

- Ordinal data only introduces a ordering/ranking (e.g. A>B), but we don't know how large A or B is.
- Cardinal further tell 'how many' of each thing, or showing quantity (e.g. 2>1)

When comparing huge numbers, I don't need to know 'how many' they are, I just need to where they are (their ranking). So I can use log to shrink their coordinates:
