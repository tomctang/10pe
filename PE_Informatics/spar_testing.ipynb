```python
# Problem D
a=int(input(''))
b=[]
for i in range(a):
    b.append(input(''))
for i in range(len(b)):
    sentence=b[i]
    sentence=sentence.split()
    for num in range(len(sentence)):
        word=sentence[num]
        word=list(map(str,word))
        if len(word)>3:
            word.reverse()
            word1=word[0]
            word[0]=word[-1]
            word[-1]=word1
        word=''.join(word)
        sentence[num]=word
    for j in range(len(sentence)-1):
        sentence[j]=sentence[j]+' '
    sentence=''.join(sentence)
    b[i]=sentence
for sentence in b:
    print(sentence)
```

```python
# Problem J
packages=int(input(''))
positions=input('')
positions=positions.split()
side=[]
bay=[]
pos_o=1
pos_b=1
sec=0
package=0
for i in range(len(positions)):
    if i%2==0:
        side.append(positions[i])
    else:
        bay.append(int(positions[i]))
o_bay=[]
b_bay=[]
o_idx=[]
b_idx=[]
print(side)
for i in range(packages):
    if side[i]=='O':
        o_bay.append(bay[i])
        o_idx.append(i)
    elif side[i]=='B':
        b_bay.append(bay[i])
        b_idx.append(i)
while packages>0:
    time_o=abs(o_bay[0]-pos_o)
    time_b=abs(b_bay[0]-pos_b)
    if side[0]=='O':
        sec+=time_o
        o_bay=o_bay[1:]
        o_idx=o_idx[1:]
        if time_b<=time_o:
            pos_b=b_bay[0]
        elif time_o<time_b:
            if b_bay[0]>pos_b:
                pos_b+=time_o
            elif b_bay[0]<pos_b:
                pos_b-=time_o
    elif side[0]=='B':
        sec+=time_b
        b_bay=b_bay[1:]
        b_idx=b_idx[1:]
        if time_o<=time_b:
            pos_o=o_bay[0]
        elif time_b<time_o:
            if o_bay[0]>pos_o:
                pos_o+=time_b
            elif o_bay[0]<pos_o:
                pos_o-=time_b
    packages-=1
    package+=1
    sec+=1
    side=side[1:]
    bay=bay[1:]
print(sec)
```
